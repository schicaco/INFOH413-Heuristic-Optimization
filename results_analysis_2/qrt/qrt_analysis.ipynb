```python
import pandas as pd
import numpy as np 

def compute_stats(file_path):
    df = pd.read_csv(file_path)

    # Detect columns automatically
    runtime_col = [c for c in df.columns if "time" in c.lower()][0]
    algo_col = [c for c in df.columns if "algo" in c.lower()][0]

    # Define success: reached target (not -1)
    df["success"] = df[runtime_col] != -1

    # Compute statistics
    stats = df.groupby(algo_col).agg(
        mean_time=(runtime_col, "mean"),
        median_time=(runtime_col, "median"),
        std_time=(runtime_col, "std"),
        min_time=(runtime_col, "min"),
        max_time=(runtime_col, "max"),
        success_count=("success", "sum"),
        total_runs=("success", "count")
    ).reset_index()

    # Compute success rate
    stats["success_rate_%"] = (stats["success_count"] / stats["total_runs"]) * 100

    return stats


# 📁 Files
file1 = "qrt_results_stabu2.csv"
file2 = "qrt_results_t70l11xx_150.csv"

# 📊 Compute tables
table1 = compute_stats(file1)
table2 = compute_stats(file2)

# Print nicely
print("\n=== STABU2 ===")
table1 = table1[["algorithm", "mean_time", "min_time", "max_time", "success_rate_%"]]

table1.loc[0, "algorithm"] = "RII"
table1.loc[1, "algorithm"] = "VNS"

table1.loc[1, "mean_time"] = np.nan
table1.loc[1, "min_time"] = 3.675639
table1.loc[1, "max_time"] = np.nan


table1


# Optional: save to CSV

table1.to_csv("stats_stabu2.csv", index=False)
```
